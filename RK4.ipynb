# Runge-Kutta 4th Order Method (RK4)

Simple exercise solving EDO with initial conditions using Runge-Kutta 4th (RK-4) order method

$\begin{array}{lcl} \displaystyle \frac{dx}{dy} & = & xy \\ y(0) & = & 2 \end{array}$

Analitical Solution: $y = 2e^{x^{2}/2}$

```
#load libraries
import numpy as np
import matplotlib.pyplot as plt
```

```
##### INPUT ###############
dx   = 0.005
xmax = 1.0
x_sol = np.arange(0.0,xmax+dx,dx)

y_sol = 2.0*np.exp(0.5*(x_sol**2))

# RK4 PARAMETERS
h = 0.25
n_iter = 11

f = lambda x,y: x*y

# CB o CI
x0 = 0.0
y0 = 2.0


x = np.zeros([n_iter,1])
y = np.zeros([n_iter,1])
x[0] = x0
y[0] = y0
```

```
for i in range(n_iter-1):

	k1 = f(x[i],y[i])
	k2 = f(x[i]+0.5*h,y[i]+0.5*h*k1)
	k3 = f(x[i]+0.5*h,y[i]+0.5*h*k2)
	k4 = f(x[i]+h,y[i]+h*k3)

	y[i+1] = y[i] + (h/6.0)*(k1+2.0*k2+2.0*k3+k4)
	x[i+1] = x[i]+h
```

```
plt.figure(figsize=(10,10))
plt.plot(x_sol,y_sol,'b',label='Analitical')
plt.plot(x,y,'r',label='Numerical')
plt.grid(color='lightgray', linestyle='--', linewidth=0.5)
plt.legend(fontsize=15)
plt.xlabel('$x$',fontsize=15)
plt.ylabel('$y$',fontsize=15)
plt.xlim(0,1)
plt.ylim(1.95,3.5)
plt.title('Runge-Kutta 4th order (RK-4) - h='+str(h),fontsize=12.5)
#plt.text(0.15,2.5,r'$\displaystyle \frac{dx}{dy} = xy$ \\ $y_{0}(x_{0} = 0)=2$',fontsize=15)
#plt.text(0.15,2.5,r'$\begin{array}{lcl} \displaystyle \frac{dx}{dy} & = & xy \\ y(0) & = & 2 \end{array}$',fontsize=15)
plt.show()
```
